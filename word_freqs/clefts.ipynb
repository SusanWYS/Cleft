## Finding Word Frequencies

### Access Google Ngram by Building URL Requests

```python
import requests 
import urllib 
import numpy as np
import json
import pandas as pd
```

```python
def runQuery(query, start_year=2000, 
            end_year=2019, corpus=26, 
            smoothing=0): 

    # converting a regular string to 
    # the standard URL format 
    # eg: "geeks for,geeks" will 
    # convert to "geeks%20for%2Cgeeks" 
    query = urllib.parse.quote(query) 

    # creating the URL 
    url = 'https://books.google.com/ngrams/json?content=' + query + \
    '&year_start=' + str(start_year) + '&year_end=' + \
    str(end_year) + '&corpus=' + str(corpus) + '&smoothing=' + \
    str(smoothing) + '' 

    # requesting data from the above url 
    response = requests.get(url) 

    # extracting the json data from the response we got 
    output = response.json() 

    # creating a list to store the ngram data 
    return_data = [] 

    if len(output) == 0: 
        # if no data returned from site, 
        # print the following statement 
        return "No data available for this Ngram."
    else: 
        # if data returned from site, 
        # store the data in return_data list 
        for num in range(len(output)): 
            
            # getting the name 
            return_data.append((output[num]['ngram'], 
                                
                                # getting ngram data 
                                output[num]['timeseries']) 
                            ) 

    return return_data 
```

```python
def word_freq(words):
    """
    Calculate the average frequency of each queried entry in a nested list from 2000 to 2019
    Args:
        words (2d list): a nested list of words to be queried (monograms or bigrams)
    Return:
        words_freq (2d list): the avg frequency of the words (among all monograms or bigrams)
    """
    words_freq = [[np.mean(runQuery(ww)[0][1]) for ww in w] for w in words]
    return words_freq
```

```python
def row_mean(lis):
    """
    Calculate the mean per row given a 2d nested list
    """
    return [np.mean(row) for row in lis]
```

```python
def create_df(verbs_freqs,VP_freqs,VP,prep):    
    verbs_meanfreqs = np.asarray(row_mean(verbs_freqs))

    col1 = np.hstack((verbs_meanfreqs,np.mean(verbs_meanfreqs)))

    VP_meanfreqs = np.asarray(row_mean(VP_freqs))

    col2 = np.hstack((VP_meanfreqs,np.mean(VP_meanfreqs)))

    col3 = np.hstack((VP_meanfreqs/verbs_meanfreqs,np.mean(VP_meanfreqs/verbs_meanfreqs)))

    data = np.vstack((col1,col2,col3))

    index = np.asarray(["\""+w[0]+"\"" for w in VP] + ["Mean Frequency"])

    columns = np.asarray(["[verb] *",f"{prep}",f"{prep}/[verb] *"])

    return pd.DataFrame(data = data.T, index = index, columns = columns)
```
